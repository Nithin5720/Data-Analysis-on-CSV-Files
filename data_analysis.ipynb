```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load CSV file
data = pd.read_csv("sales_data.csv")

# Step 3: Display first few rows
print("üìä Sample Data:")
print(data.head())

# Step 4: Add a 'Total_Sales' column
data["Total_Sales"] = data["Quantity"] * data["Price"]

# Step 5: Basic Insights
print("\n‚úÖ Basic Information:")
print(data.info())

print("\nüìà Summary Statistics:")
print(data.describe())

# Step 6: Total sales by product
sales_by_product = data.groupby("Product")["Total_Sales"].sum()
print("\nüí∞ Total Sales by Product:")
print(sales_by_product)

# Step 7: Total sales by region
sales_by_region = data.groupby("Region")["Total_Sales"].sum()
print("\nüåç Total Sales by Region:")
print(sales_by_region)

# Step 8: Plot the charts
plt.figure(figsize=(8,4))
sales_by_product.plot(kind='bar', color='skyblue')
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales (‚Çπ)")
plt.grid(True)
plt.show()

plt.figure(figsize=(8,4))
sales_by_region.plot(kind='bar', color='orange')
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales (‚Çπ)")
plt.grid(True)
plt.show()
```
